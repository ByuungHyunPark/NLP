```python
import re
import nltk
from konlpy.tag import Kkma

def engTitle(mainTitle): #영어 제목 전처리
    
    nouns = []
    for word in nltk.tokenize.word_tokenize(mainTitle):
        word = word.lower() #소문자 변환
        if word not in nltk.corpus.stopwords.words('english'): #불용어 제거
            nouns.append(word)

    return nouns



def titleKeyword(mainTitle): #논문 제목 전처리

    mainTitle = re.sub(pattern="[^ a-zA-Z가-힣]", repl='', string=mainTitle) #특수문자 제거

    try:
        kkma = Kkma()   # 형태소 분석기Kkma
        kkma_pos = kkma.pos(mainTitle) #형태소 추출

        if len(kkma.nouns(mainTitle)) == 0: #한글 이 없는 영문 제목 데이터 정제
            nouns = engTitle(mainTitle)

        else: #한글논문 데이터 Konlpy(kkma)를 통한 명사 추출
            nouns = [n for n, tag in kkma_pos if tag in ['NNG', 'NNP', 'OL', 'UN']]

    except:
        nouns = []
        pass
    
    return nouns
```
